```python
data = pd.read_csv("cost_revenue_dirty.csv")
```

```python
X = DataFrame(data, columns=['production_budget'])
y = DataFrame(data, columns=['worldwide_gross'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

<h2>CREATE LinearRegression</h2>

```python
# Create linear regression object
regression = LinearRegression()

# Train the model using the training sets
regression.fit(X, y)

```

```python
# The coefficients / m / theta_1
print('Coefficients: \n', regression.coef_)
```

<i>It means that there is a positive relationship between budget and revenue and not only that,<br>
it means that for each dollar that we spend on producing the movie, we should get around 3.1<br> 
dollars in revenue in return. </i>

```python
# the intercept / c / theta_0
print('Intercept: \n', regression.intercept_)
```

<i>over 7 million dollars. So that's a bit problematic. Right? That seems quite unrealistic because if you and I went <br> out to make a movie with a thousand dollars, it's pretty unlikely that seven million would just <br> disappear from our bank accounts. So, this is a lot less realistic. </i>

```python
# Plot outputs
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.plot(X, regression.predict(X), color='red', linewidth=4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

<h2>PREDICT YOUR REVENUE</h2>

```python
m = regression.coef_[0][0]
c = regression.intercept_[0]


print("The equation of the regression line is y = {m}x + {c}".format(m=m, c=c)) 

print("Enter the production budget of the film to predict the revenue")

budget = float(input())

y_val = m * budget + c

print("The predicted revenue of the film is {y_val}".format(y_val=y_val))
```

```python
regression.score(X,y)
```
