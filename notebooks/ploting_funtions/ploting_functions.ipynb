```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### FUNCTION
$f(x) = x^2+x+1$
<br>
<br>
$f'(x) = 2x+1$
<br>
<P> PLOT FUNCTIONS WITH NUMPY </p>

```python
def f(x):
    return x**2+x+1

def df(x):
    return 2*x+1
```

```python
#make values for x

x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
#plot the function and derivative
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('f(x) = x^2+2x+1')
plt.grid()
plt.plot(x_1, f(x_1), color='lightblue', linewidth=3, label='f(x)')
plt.subplot(1,2,2)
plt.grid()
plt.title('df(x) = 2x+2')
plt.plot(x_1, df(x_1), color='seagreen', linewidth=3, label='df(x)')
```

```python
#GRADIENT DESCENT
#initialize variables

new_x = 3
previous_x = 0
precision = 0.00001
gamma = 0.1

x_val = [new_x]
slope_val = [df(new_x)]

for i in range(500):
    previous_x = new_x
    new_x = previous_x - gamma*df(previous_x)
    step_size = abs(new_x - previous_x)

    x_val.append(new_x)
    slope_val.append(df(new_x))

    if step_size < precision:
        print('Number of steps: ', i)
        break

x_val = np.array(x_val)
slope_val = np.array(slope_val)

print('Local minimum occurs at: ', new_x)
print("slope or df(x) value is: ", df(new_x))
print('Minimum value is or cost is: ', f(new_x))
print('slope values: ', slope_val)
print('x values: ', x_val)
```

```python
#plot the function and derivative
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('cost function f(x) = x^2+2x+1')
plt.grid()
plt.plot(x_1, f(x_1), color='lightblue', linewidth=3, label='f(x)')
plt.scatter(x_val, f(np.array(x_val)), color='red', s=100, alpha=0.6)

#plot slove function
plt.subplot(1,3,2)
plt.grid()
plt.title('slope funtion df(x) = 2x+2')
plt.plot(x_1, df(x_1), color='seagreen', linewidth=3, label='df(x)',alpha=0.6)
plt.scatter(x_val, slope_val, color='red', s=100, alpha=0.6)

#plot the gradient descent clsoeup
plt.subplot(1,3,3)
plt.xlim([new_x-0.2, new_x+0.2])
plt.grid()
plt.title('Gradient Descent (closeup)')
plt.plot(x_1, df(x_1), color='seagreen', linewidth=3, label='df(x)',alpha=0.6)
plt.scatter(x_val, slope_val, color='red', s=100, alpha=0.6)

```
